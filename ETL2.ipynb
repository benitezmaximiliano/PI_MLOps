```python

df_user_items = df_user_items[['User_Id','Item_Id','Playtime_Forever']]
df_user_items['Item_Id'] = df_user_items['Item_Id'].astype(str)
df_user_items['Id'] = df_user_items['User_Id'] + df_user_items['Item_Id']

df_steam_games = df_steam_games.rename(columns={'Id': 'Item_Id'})
df_steam_games['Item_Id'] = df_steam_games['Item_Id'].astype(str)

df_user_reviews['Item_Id'] = df_user_reviews['Item_Id'].astype(str)
df_user_reviews['Id'] = df_user_reviews['User_Id'] + df_user_reviews['Item_Id']
```

```python
merged_df = df_user_reviews.merge(df_steam_games, on='Item_Id', how='left')
merged_df.dropna(inplace=True)
#Generamos dataframe definitivo uniendo 'items' con 'merge_df' por el identificador unico 'id' de nombre "steam"
steam = df_user_items.merge(merged_df, on='Id')
#Renombramos columnas del dataframe 'steam'
steam = steam.rename(columns={'User_Id_x':'User_Id'})
steam = steam.rename(columns={'Item_Id_x':'Item_Id'})
#Eliminamos las columnas redundantes del dataframe 'steam'
steam.drop('User_Id_y', axis='columns', inplace=True)
steam.drop('Item_Id_y', axis='columns', inplace=True)
```

```python
steam
```

```python
steam['Posted'] = steam['Posted'].astype('str')
```

```python
#steam['Posted'] = steam['Posted'].astype('int64')
steam['Playtime_Forever'] = steam['Playtime_Forever'].astype('int64')
steam['Year'] = steam['Year'].astype('int64')
```

```python
steam = steam[['User_Id','Item_Id','Playtime_Forever','Recommend','Sentiment_Score','Posted','Genres','Title','Developer','Year']]
```
